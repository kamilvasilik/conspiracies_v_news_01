# Exploratory Data Analysis (EDA)

### Conspiracy sites

```python
import pandas as pd
```

```python
data_dtm = pd.read_pickle('dtm.pkl')
data_dtm = data_dtm.transpose()
data_dtm.head()
```

```python
top_dict = {}
for c in data_dtm.columns:
    top = data_dtm[c].sort_values(ascending=False).head(30)
    top_dict[c] = list(zip(top.index, top.values))
    
top_dict
```

```python
from collections import Counter
```

```python
words = []
for art in data_dtm.columns:
    top = [word for (word, count) in top_dict[art]]
    for t in top:
        words.append(t)
words
```

```python
Counter(words).most_common()
```

## WordCloud

```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
```

```python
data_clean = pd.read_pickle('data_cleaner.pkl')
```

```python
from sklearn.feature_extraction import text
```

```python
stop_words = text.ENGLISH_STOP_WORDS
stop_words
```

```python
wc = WordCloud(stopwords=stop_words, background_color='white', colormap='Dark2', max_font_size=150, random_state=42)
```

```python
plt.rcParams['figure.figsize'] = [16, 6]

for index, theme in enumerate(data_dtm.columns):
    wc.generate(data_clean.article[theme])
    
    plt.subplot(2, 5, index+1)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(theme)
    
plt.show()
```
